## 2. Configurazione dei Segreti

Crea un file chiamato **`.env`** (attenzione al punto iniziale e niente estensioni come .txt).

**Regole d'oro:**

* Nomi in **MAIUSCOLO**.
* Niente spazi attorno all'uguale `=`.
* Un segreto per ogni riga.

```text
DROPBOX_TOKEN=sl.uYtg67your_secret_token_here
DATABASE_URL=localhost:5432

```

## 3. Utilizzo nello Script Python

Inizia ogni script con questi import:

```python
import os
from dotenv import load_dotenv
from pathlib import Path

```

### SCENARIO A: Il file `.env` Ã¨ nella stessa cartella dello script

Usa questo metodo se tieni i segreti dentro la cartella specifica della sezione (es: `Sezione 08/.env`).

```python
# Carica automaticamente il file .env presente nella cartella corrente
load_dotenv()

# Recupera la chiave
token = os.getenv("DROPBOX_TOKEN")

```

### SCENARIO B: Il file `.env` Ã¨ nella Root principale (Consigliato)

Usa questo metodo se hai un unico file `.env` per tutto il corso, situato nella cartella principale (`LEARN_PYTHON/.env`).

```python
# 1. Trova la cartella dove si trova lo script (es: Sezione 08)
cartella_script = Path(__file__).resolve().parent

# 2. Risale alla cartella Root (il genitore della cartella script)
root_progetto = cartella_script.parent 

# 3. Costruisce il percorso del file .env e lo carica
path_env = root_progetto / '.env'
load_dotenv(dotenv_path=path_env)

# 4. Recupera la chiave
token = os.getenv("DROPBOX_TOKEN")

```
